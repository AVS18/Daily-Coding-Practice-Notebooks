# Question 1

Given an Input array, arr. you have to produce an ouptut such that output[i] is equal to product of all elements in the array except arr[i]<br>
Input: 1,2,3,4<br>
Output: 24,12,8,6<br>
You should not use division and complete in O(n) time<br>

```python
#gfg copied
arr=list(map(int,input().split()))
z=0
l_r=[0]*len(arr)
r_r=[0]*len(arr)
out=[0]*len(arr)
l_r[0]=1
r_r[-1]=1
for i in range(1,len(l_r)):
    l_r[i]=l_r[i-1]*arr[i-1]
for j in range(len(arr)-2,-1,-1):
    r_r[j]=r_r[j+1]*arr[j+1]
for i in range(0,len(arr)):
    out[i]=l_r[i]*r_r[i]
print(out)
```
